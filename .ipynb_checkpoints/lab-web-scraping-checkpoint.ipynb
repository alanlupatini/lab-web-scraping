```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re # For regular expressions, useful for parsing data like price and rating

BASE_URL = "https://books.toscrape.com/"


def get_book_details(book_url):
    """
    Scrapes the detail page
    
    Args:
        book_url (str): The absolute URL of the book's detail page.

    Returns:
        UPC, Genre, Description
    """
    try:
        response = requests.get(book_url)
        response.raise_for_status() # Check for bad status codes
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # 1. Extract Description
        description_tag = soup.find('div', id='product_description')
        description = description_tag.find_next_sibling('p').text if description_tag else 'N/A'
        
        # 2. Extract UPC and other Product Information
        product_table = soup.find('table', class_='table table-striped')
        
        upc = 'N/A'
        genre = 'N/A'
        
        if product_table:
            upc = product_table.find_all('td')[0].text
        
        # 3. Extract Genre from the breadcrumba
        breadcrumb = soup.find('ul', class_='breadcrumb')
        # The genre is the third <li>
        genre_tag = breadcrumb.find_all('li')[2] if breadcrumb else None
        genre = genre_tag.text.strip() if genre_tag else 'N/A'
        
        return upc, genre, description
        
    except Exception as e:
        print(f"Error scraping details for {book_url}: {e}")
        return 'N/A', 'N/A', 'N/A'

# HERE GOES THE SCRAPEEEEEEEEEE

def scrape_books(min_rating: int, max_price: float) -> pd.DataFrame:
    """
    Scrapes book data from Books to Scrape, handles pagination, 
    and filters results based on rating and price.

    Args:
        min_rating (int): Minimum star rating (1-5) to include.
        max_price (float): Maximum price in Pounds (Â£) to include.

    Returns:
        pd.DataFrame: A DataFrame of filtered book data.
    """
    all_books_data = []
    page_num = 1
    

    while True:
        if page_num == 1:
            url = BASE_URL + 'index.html'
        else:
            url = BASE_URL + f'catalogue/page-{page_num}.html'
            

        try:
            response = requests.get(url)
            response.raise_for_status() 
        except requests.exceptions.HTTPError:
            print("Reached the last page or encountered an error. Stopping pagination.")
            break
        except requests.exceptions.RequestException as e:
            print(f"An error occurred while fetching {url}: {e}")
            break

        soup = BeautifulSoup(response.content, 'html.parser')
        
        book_articles = soup.find_all('article', class_='product_pod')
        
        if not book_articles:
            print("No more books found. Stopping pagination.")
            break

        for article in book_articles:
            
            # 1. Title and URL
            title_tag = article.find('h3').find('a')
            title = title_tag['title']
            relative_url = title_tag['href']

            if 'catalogue' in relative_url:
                 full_url = BASE_URL + relative_url.replace('../', '')
            else:
                 full_url = BASE_URL + 'catalogue/' + relative_url
            
            # 2. Price
            price_str = article.find('p', class_='price_color').text.strip()
            # Use regex to extract the number and convert to float
            price_match = re.search(r'[\d\.]+', price_str)
            price = float(price_match.group(0)) if price_match else 0.0
            
            # 3. Rating
            rating_str = article.find('p', class_=re.compile(r'star-rating'))['class'][1]
            # Convert text rating (e.g., 'Two') to an integer (2)
            rating_map = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
            rating = rating_map.get(rating_str, 0)
            
            # 4. Availability
            availability = article.find('p', class_='instock').text.strip()
            
            if rating >= min_rating and price <= max_price:
                upc, genre, description = get_book_details(full_url)
                
                book_data = {
                    'UPC': upc,
                    'Title': title,
                    'Price (Â£)': price,
                    'Rating': rating,
                    'Genre': genre,
                    'Availability': availability,
                    'Description': description
                }
                all_books_data.append(book_data)
                
        page_num += 1

    print(f"\nFinished scraping. Found {len(all_books_data)} books matching criteria before final DataFrame creation.")
    
    df = pd.DataFrame(all_books_data)
    
    return df


if __name__ == '__main__':
    MIN_RATING_THRESHOLD = 4
    MAX_PRICE_THRESHOLD = 20.0
    
    try:
        final_df = scrape_books(
            min_rating=MIN_RATING_THRESHOLD, 
            max_price=MAX_PRICE_THRESHOLD
        )
        
        print("\n" + "="*50)
        print("Filtered Books (min_rating >= 4, max_price <= 20)")
        print("="*50)
        
        if not final_df.empty:
            print(f"Total books found: {len(final_df)}")
            # Display the first 10 results and the column types
            print("\nFirst 10:")
            print(final_df.head(10).to_markdown(index=False))
            print("\nDF Info:")
            final_df.info()
            
        else:
            print("No books found.")

    except Exception as e:
        print(f"\nERROR: {e}")
```

```python
final_df.to_csv('highly_rated_books.csv', index=False)
print("\nsaved to highly_rated_books.csv")
```
